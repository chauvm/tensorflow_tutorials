```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array(np.arange(-10.0, 10.0, 0.01))
```

# Plot Step function

   $$ y = \begin{cases}
      1, & \text{if}\ t > 0 \\
      0, & \text{otherwise}
    \end{cases}$$
    
   $$ y = \begin{cases}
      0.5, & \text{if}\ t > 0 \\
      -0.5, & \text{otherwise}
    \end{cases}$$

```python
Y = X > 0 
Y = Y - 0.5
plt.plot(X, Y)
plt.xlabel("t")
plt.ylabel("y = 0.5 if t > 0 else -0.5")
# add horizontal and vertical lines through origin
plt.axhline(y=0, color='k')
plt.show()
```

# Plot sigmoid
$$S(t) = \frac{1}{1 + exp(-t)}$$

```python
def sigmoid(s):
    return 1 / (1.0 + np.exp(-s))
```

```python
Y = sigmoid(X)
```

```python
plt.plot(X, Y)
plt.xlabel("t")
plt.ylabel("S(t) = 1/(1 + exp(-t))")
# add horizontal and vertical lines through origin
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

# Plot tanH

$$tanh(t) = \frac{2}{1 + exp(-t)} - 1$$

```python
def tanh(x):
    return 2*sigmoid(x) - 1
```

```python
Y_tanh = tanh(X)
```

```python
plt.plot(X, Y_tanh)
plt.xlabel("t")
plt.ylabel("S(t) = 2/(1 + exp(-t)) - 1")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

# Plot ReLU
$$ReLU(t) = max(0, t)$$

```python
def relu(x):
    return np.maximum(0, x)
```

```python
Y_relu = relu(X)
```

```python
plt.plot(X, Y_relu)
plt.xlabel("t")
plt.ylabel("ReLU(t) = max(0, t)")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```
